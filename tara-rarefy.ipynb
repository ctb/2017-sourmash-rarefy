````python
labels = { 'IO': 'Indian Ocean',
          'MS': 'Mediterranean',
          'NAO': 'North Atlantic',
          'NPO': 'North Pacific Ocean',
          'RS': 'Red Sea',
          'SAO': 'South Atlantic Ocean',
          'SO': 'Southern Ocean',
          'SPO': 'South Pacific Ocean'}
````

````python
import pandas

d = {}
for i in ['IO', 'MS', 'NAO', 'NPO', 'RS', 'SAO', 'SO', 'SPO']:
    d[i] = pandas.read_csv('tara-100k-{}.csv'.format(i))
````

````python
%matplotlib inline
from matplotlib import pyplot
````

````python
pyplot.figure(figsize=(10, 6))
for k in d:
    df = d[k]
    pyplot.plot(df.n*100000 / 1e9, df.k * 100000 / 1e9, label=labels[k])
    
pyplot.legend(loc='lower right')
pyplot.xlabel('number of words (billion)')
pyplot.ylabel('distinct words (billion)')
pyplot.savefig('/tmp/rarefy-all.pdf')
````

````python
pyplot.figure(figsize=(10, 6))
for k in d:
    df = d[k]
    pyplot.plot(df.n, df.known, label=labels[k])
    
pyplot.legend(loc='lower right')
pyplot.xlabel('number of hashes')
pyplot.ylabel('distinct hashes')
pyplot.title('known hashes only')
````

````python
pyplot.figure(figsize=(10, 6))
for k in d:
    df = d[k]
    pyplot.plot(df.n, df.k, label=labels[k])
    
pyplot.legend(loc='upper left')
pyplot.xlabel('number of hashes')
pyplot.ylabel('distinct hashes')
pyplot.title('zoom - all hashes, known & unknown')
pyplot.axis(xmax=100000, ymax=100000)
````

````python
pyplot.figure(figsize=(10, 6))
for k in d:
    df = d[k]
    pyplot.plot(df.n, df.known, label=labels[k])
    
pyplot.legend(loc='upper left')
pyplot.xlabel('number of hashes')
pyplot.ylabel('distinct hashes')
pyplot.title('zoom - known hashes only')
pyplot.axis(xmax=120000, ymax=2000)
````
